```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (adjust the file path if needed)
#from google.colab import files
#ploaded = files.upload()

# Read the uploaded CSV file
#file_name = list(uploaded.keys())[0]
data = pd.read_csv("d.csv")

# Convert "Received Time" to datetime
data['Received Time'] = pd.to_datetime(data['Received Time'], errors='coerce')

# Email Volume: Classify as Inbound or Outbound
organization_domain = "o365droid.onmicrosoft.com"
data['Direction'] = data.apply(
    lambda row: 'Inbound' if organization_domain not in row['Sender Address'] else 'Outbound', axis=1
)
email_volume = data['Direction'].value_counts()

# Top Senders and Recipients
top_senders = data['Sender Address'].value_counts().head(5)
top_recipients = data['Recipient Address'].value_counts().head(5)

# Email Size Distribution: Clean and convert "Size"
data['Size (KB)'] = pd.to_numeric(data['Size'].str.replace(' KB', ''), errors='coerce')

# Status Breakdown
status_breakdown = data['Status'].value_counts()

# Hourly and Daily Trends
data['Hour'] = data['Received Time'].dt.hour
data['Date'] = data['Received Time'].dt.date
hourly_trends = data.groupby('Hour').size()
daily_trends = data.groupby('Date').size()

# Visualization 1: Email Volume (Inbound vs. Outbound)
plt.figure(figsize=(6, 4))
email_volume.plot(kind='bar', color='skyblue')
plt.title("Email Volume (Inbound vs. Outbound)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

# Visualization 2: Top Senders
plt.figure(figsize=(6, 4))
top_senders.plot(kind='bar', color='orange')
plt.title("Top 5 Senders")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.show()

# Visualization 3: Top Recipients
plt.figure(figsize=(6, 4))
top_recipients.plot(kind='bar', color='green')
plt.title("Top 5 Recipients")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.show()

# Visualization 4: Email Size Distribution
plt.figure(figsize=(8, 5))
plt.hist(data['Size (KB)'].dropna(), bins=30, color='purple', edgecolor='black')
plt.title("Email Size Distribution")
plt.xlabel("Size (KB)")
plt.ylabel("Frequency")
plt.show()

# Visualization 5: Status Breakdown
plt.figure(figsize=(6, 6))
status_breakdown.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='coolwarm')
plt.title("Email Status Breakdown")
plt.ylabel("")  # Remove default ylabel
plt.show()

# Visualization 6: Hourly Trends
plt.figure(figsize=(8, 5))
hourly_trends.plot(kind='line', marker='o', color='blue')
plt.title("Hourly Email Activity")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Emails")
plt.grid(True)
plt.show()

# Visualization 7: Daily Trends
plt.figure(figsize=(10, 6))
daily_trends.plot(kind='line', marker='o', color='red')
plt.title("Daily Email Activity")
plt.xlabel("Date")
plt.ylabel("Number of Emails")
plt.grid(True)
plt.show()
```

```python
# Additional analysis for the email data

# 1. Average Email Size per Direction
avg_size_per_direction = data.groupby('Direction')['Size (KB)'].mean()

# 2. Top 5 Domains for Recipients (excluding organization's own domain)
data['Recipient Domain'] = data['Recipient Address'].apply(lambda x: x.split('@')[-1])
external_domains = data[data['Recipient Domain'] != organization_domain]
top_recipient_domains = external_domains['Recipient Domain'].value_counts().head(5)

# 3. Time Series of Status Changes
status_trends = data.groupby(['Date', 'Status']).size().unstack(fill_value=0)

# 4. Weekly Trends
data['Weekday'] = data['Received Time'].dt.day_name()
weekly_trends = data.groupby('Weekday').size().reindex(
    ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
)

# 5. Average Email Size by Day and Hour
avg_size_by_hour = data.groupby('Hour')['Size (KB)'].mean()
avg_size_by_day = data.groupby('Weekday')['Size (KB)'].mean().reindex(
    ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
)

# Displaying the findings
avg_size_per_direction, top_recipient_domains, status_trends.head(), weekly_trends, avg_size_by_hour.head(), avg_size_by_day.head()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# Load the data from the CSV file
url = 'd.csv'
data = pd.read_csv(url)

# Display the first few rows of the dataset
print(data.head())

# Data Cleaning: Convert 'Received Time' to datetime format
data['Received Time'] = pd.to_datetime(data['Received Time'])

# Handle Size column: Replace ' KB' and convert to numeric, handling non-convertible values
def convert_size(size):
    if 'KB' in size:
        return float(size.replace(' KB', ''))
    elif 'Byte' in size:
        return 0.0  # Treat '0 Byte' as 0 KB
    else:
        return None  # Return None for any other unexpected formats

data['Size'] = data['Size'].apply(convert_size)

# Drop rows with NaN sizes after conversion (if any)
data = data.dropna(subset=['Size'])

# Basic statistics of the dataset
stats = data.describe()
print("\nBasic Statistics of the Dataset:")
print(stats)

# Count unique statuses
status_counts = data['Status'].value_counts()
print("\nEmail Status Counts:")
print(status_counts)

# Plotting the status distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Status', data=data, palette='viridis')
plt.title('Email Status Distribution')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Identify frequent senders and recipients
frequent_senders = data['Sender Address'].value_counts().head(10)
frequent_recipients = data['Recipient Address'].value_counts().head(10)

# Plotting frequent senders
plt.figure(figsize=(10, 6))
sns.barplot(x=frequent_senders.values, y=frequent_senders.index, palette='Blues_d')
plt.title('Top 10 Frequent Senders')
plt.xlabel('Count')
plt.ylabel('Sender Address')
plt.show()

# Plotting frequent recipients
plt.figure(figsize=(10, 6))
sns.barplot(x=frequent_recipients.values, y=frequent_recipients.index, palette='Reds_d', legend=False)
plt.title('Top 10 Frequent Recipients')
plt.xlabel('Count')
plt.ylabel('Recipient Address')
plt.show()

# Check for anomalies: Emails with status 'Failed'
failed_emails = data[data['Status'] == 'Failed']
print("\nFailed Emails:")
print(failed_emails[['Received Time', 'Sender Address', 'Recipient Address', 'Subject', 'Size']])

# Analyze email sizes for potential anomalies (e.g., unusually large emails)
large_emails = data[data['Size'] > 50]  # Assuming anything over 50 KB is large
print("\nUnusually Large Emails:")
print(large_emails[['Received Time', 'Sender Address', 'Recipient Address', 'Subject', 'Size']])

# Plotting email sizes distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Size'], bins=30, kde=True)
plt.title('Email Size Distribution')
plt.xlabel('Size (KB)')
plt.ylabel('Frequency')
plt.show()

# Timing analysis: Count of emails sent per hour
data['Hour'] = data['Received Time'].dt.hour
hourly_counts = data['Hour'].value_counts().sort_index()

# Plotting hourly email counts
plt.figure(figsize=(10, 6))
sns.lineplot(x=hourly_counts.index, y=hourly_counts.values)
plt.title('Emails Sent Per Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Emails Sent')
plt.xticks(range(24))
plt.show()

# Create a summary table for analysis results
summary_table = pd.DataFrame({
    "Metric": ["Total Emails", "Failed Emails", "Unusually Large Emails"],
    "Count": [len(data), len(failed_emails), len(large_emails)]
})

print("\nSummary Table of Analysis Results:")
print(summary_table)

# Save the cleaned data for further analysis if needed
data.to_csv('cleaned_email_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# Load the data from the CSV file
url = 'd.csv'
data = pd.read_csv(url)

# Display the first few rows of the dataset
print(data.head())

# Data Cleaning
def clean_data(data):
    # Convert 'Received Time' to datetime format
    data['Received Time'] = pd.to_datetime(data['Received Time'])

    # Handle Size column: Replace ' KB', ' Byte' and convert to numeric
    def convert_size(size):
        if 'KB' in size:
            return float(size.replace(' KB', '').strip())
        elif 'Byte' in size:
            return 0.0  # Treat '0 Byte' as 0 KB
        return None  # Return None for other unexpected formats

    data['Size'] = data['Size'].apply(convert_size)

    # Drop rows with NaN sizes after conversion
    return data.dropna(subset=['Size'])

# Clean the dataset
data = clean_data(data)

# Function to plot top senders
def plot_top_senders(data):
    high_volume_senders = data['Sender Address'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=high_volume_senders.values, y=high_volume_senders.index, palette='Blues_d')
    plt.title('Top 10 High-Volume Senders')
    plt.xlabel('Number of Emails Sent')
    plt.ylabel('Sender Address')
    plt.show()

# Plot high-volume senders
plot_top_senders(data)



# Function to plot large emails
def plot_large_emails(data, size_threshold=50):
    large_emails = data[data['Size'] > size_threshold]
    plt.figure(figsize=(10, 6))
    sns.histplot(large_emails['Size'], bins=30, kde=True)
    plt.title(f'Distribution of Large Emails (> {size_threshold} KB)')
    plt.xlabel('Size (KB)')
    plt.ylabel('Frequency')
    plt.show()

# Plot large emails
plot_large_emails(data)

# Function to plot external communication
def plot_external_communication(data):
    external_domains = data[~data['Recipient Address'].str.contains('@o365droid.onmicrosoft.com')]
    external_counts = external_domains['Recipient Address'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=external_counts.values, y=external_counts.index, palette='Greens_d')
    plt.title('Top 10 External Recipients')
    plt.xlabel('Number of Emails Sent to External Domains')
    plt.ylabel('Recipient Address')
    plt.show()

# Plot external communication
plot_external_communication(data)

# Function to plot non-business hour activity
def plot_non_business_hour_activity(data):
    data['Hour'] = data['Received Time'].dt.hour
    non_business_hours = data[(data['Hour'] < 9) | (data['Hour'] >= 17)]
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Hour', data=non_business_hours, palette='Purples_d')
    plt.title('Email Activity Outside Business Hours')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Emails Sent')
    plt.xticks(range(24))
    plt.show()

# Plot non-business hour activity
plot_non_business_hour_activity(data)

# Function to plot email status trends by hour of the day
def plot_status_trends(data):
    status_trends = data.groupby(['Hour', 'Status']).size().unstack(fill_value=0)
    status_trends.plot(kind='bar', stacked=True, figsize=(12, 7))
    plt.title('Email Status Trends by Hour of Day')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Emails')
    plt.xticks(range(24))
    plt.legend(title='Status', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

# Plot status trends
plot_status_trends(data)
```

```python
# Function to plot failed delivery reasons with enhancements
def plot_failed_deliveries(data):
    failed_deliveries = data[data['Status'] == 'Failed']
    failed_counts = failed_deliveries['Subject'].value_counts().head(10)  # Focus on top 10 reasons

    plt.figure(figsize=(12, 8))
    sns.barplot(x=failed_counts.values, y=failed_counts.index, palette='Reds_d', orient='h')

    # Add labels to each bar
    for index, value in enumerate(failed_counts.values):
        plt.text(value + 0.5, index, str(value), va='center')

    plt.title('Reasons for Failed Deliveries', fontsize=16)
    plt.xlabel('Count of Failed Deliveries', fontsize=12)
    plt.ylabel('Subject', fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Plot enhanced failed delivery reasons
plot_failed_deliveries(data)
```
